##### Spread of a virus

```python
import matplotlib.pyplot as plt

initial_people_infected = 5  # initial number of infected people
infection_rate = 2           # number of people infected by each infected person
rounds = 5

infected = [initial_people_infected]

for i in range(rounds):
    infected.append(infected[-1]*infection_rate)

plt.plot(infected)
plt.xlabel("Rounds of Infection")
plt.ylabel("Number of Infected People")
plt.title("Exponential Virus Spread")
plt.show()

    
```
